```python
import pandas as pd
import numpy as np
import nltk
import statsmodels.api as sm
from sklearn.preprocessing import MinMaxScaler

pd.set_option('display.max_colwidth', 200)

import matplotlib.pyplot as plt
import seaborn as sns

from nltk.sentiment.vader import SentimentIntensityAnalyzer
from sklearn.model_selection import train_test_split

import warnings 

warnings.filterwarnings("ignore")
```

```python
df_comments = pd.read_csv('data_comments_31-08-2022.csv')
df_comments = df_comments.drop(['authorProfileImageUrl', 'authorChannelUrl', 'authorChannelId', 'authorDisplayName', 'canRate', 'viewerRating', 'parentId'], axis = 1)
```

```python
df_comments.head()
```

```python
df_comments['cleanedText'] = df_comments['textOriginal'].str.strip()
clean = df_comments.pop('cleanedText')
df_comments.insert(3, 'cleanedText', clean)
df_comments.head()

from nltk.corpus import stopwords
```

```python
df_videos = pd.read_csv('data_videos_31-08-2022.csv')
df_videos.head()
```

```python

df_videos['ratio'] = df_videos['likeCount']/df_videos['viewCount']



scaler = MinMaxScaler()

x = np.array(df_videos['ratio'])

normal = scaler.fit_transform(x.reshape(-1, 1))

df_videos['normalized'] = normal

df_videos.head()
```

```python
nltk.download('vader_lexicon', quiet = True)
```

```python
pol = []
stance = []
analyzer = SentimentIntensityAnalyzer()

for text in df_comments['textDisplay']:
    score = analyzer.polarity_scores(text)['compound']
    pol.append(score)
    if score > 0.5:
        stance.append(1)
    elif score < -0.5:
        stance.append(-1)
    else:
        stance.append(0)
    
    
df_comments['polarity'] = pol
df_comments['stance'] = stance
```

```python
df_comments.head()
```

```python
df_comments = df_comments[df_comments['polarity'] != 0.0] 

print(len(df_comments.index))
df_comments.tail()
```

```python
means= []
for videoId in df_videos['id']:
    #print(videoId)
    #targets = df_comments.loc[df_comments['videoId'] == videoId, 'polarity']
    df = (df_comments).loc[(df_comments)['videoId'] == videoId]
    q3, q1 = np.percentile(np.array(df['polarity']), [87.5 ,12.5])
    df = df[(df['polarity'] >= q1) & (df['polarity'] <= q3)]
    mean = df['polarity'].mean()
    means.append(mean)
    
#print(res)

df_videos['avgPolarity'] = means   
df_videos.head(50)
    #print(df)
    #print(targets)
    #res.append(targets['polarity'].quantile(q = 0.5))

#df_video['avgPolarity'] = res
    
```

```python
y = df_videos['normalized']
x = df_videos['avgPolarity']

x = sm.add_constant(x)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

```python
linearModel = sm.OLS(y_train,X_train).fit()


linearModel.predict(X_test)


linearModel.summary()
```
